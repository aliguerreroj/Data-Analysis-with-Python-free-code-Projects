```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar los datos (simulando que ya tienes un DataFrame llamado df)
df = pd.read_csv('fcc-forum-pageviews.csv', index_col=0,parse_dates=['date'])

# 2. Crear nuevas columnas para AÃ±o y Mes
df["Year"] = df.index.year
df["Month"] = df.index.month

# 3. Agrupar los datos por AÃ±o y Mes, calculando el promedio de 'value'
monthly_avg = df.groupby(["Year", "Month"])["value"].mean().unstack()

# 4. Crear la grÃ¡fica
plt.figure(figsize=(10, 6))

# Configurar colores con Seaborn para una mejor visualizaciÃ³n
colors = sns.color_palette("tab10", 12)

# 5. Graficar los datos como barras agrupadas
monthly_avg.plot(kind="bar", figsize=(12, 7), width=0.8, color=colors)

# # 6. Configurar etiquetas y tÃ­tulo
plt.xlabel("Years", fontsize=12)
plt.ylabel("Average Page Views", fontsize=12)
plt.title("Monthly Average Page Views Over the Years", fontsize=14)
plt.legend(title="Months", labels=[
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
], fontsize=10)
plt.xticks(rotation=45)  # Rotar etiquetas del eje X

# 7. Mostrar la grÃ¡fica
plt.show()
```

```python
def draw_bar_plot():
    # Cargar datos
    df = pd.read_csv('fcc-forum-pageviews.csv', index_col=0, parse_dates=['date'])

    # Extraer aÃ±o y mes
    df['Year'] = df.index.year
    df['Month'] = df.index.month_name()

    # Calcular el promedio por aÃ±o y mes
    monthly_avg = df.groupby(['Year', 'Month'])['value'].mean().reset_index()

    # Pivotar para estructurar los datos para la grÃ¡fica
    data_pivot = monthly_avg.pivot(index='Year', columns='Month', values='value')

    # Ordenar los meses en el orden correcto
    month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
                   'July', 'August', 'September', 'October', 'November', 'December']
    data_pivot = data_pivot[month_order]

    # Configurar colores con Seaborn
    colors = sns.color_palette("tab10", 12)

    # Crear figura y ejes
    plt.close('all') # Cerrar figuras previas
    fig, ax = plt.subplots(figsize=(12, 8))

    # Graficar
    data_pivot.plot(kind='bar', ax=ax, color=colors)

    # Personalizar
    ax.set_title('Average Page Views per Month', fontsize=16)
    ax.set_xlabel('Years', fontsize=12)
    ax.set_ylabel('Average Page Views', fontsize=12)
    ax.legend(title='Months', fontsize=10)
    ax.grid(axis='y')

    # Guardar la imagen y retornar la figura
    fig.savefig('bar_plot.png')
    return fig
draw_bar_plot()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('fcc-forum-pageviews.csv', index_col=0, parse_dates=['date'])

df = df[(df['value'] >= df['value'].quantile(0.025)) & 
        (df['value'] <= df['value'].quantile(0.975))]

df.loc[:, 'Year'] = df.index.year
df.loc[:, 'Month'] = df.index.strftime('%b')

def draw_box_plot():
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    sns.boxplot(x='Year', y='value', hue='Year', data=df, ax=axes[0], palette='Set2', legend=False)
    axes[0].set_title('Year-wise Box Plot (Trend)')
    axes[0].set_xlabel('Year')
    axes[0].set_ylabel('Page Views')
    

    
    month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    sns.boxplot(x='Month', y='value', hue='Month', data=df, ax=axes[1], palette='Set3', order=month_order, legend=False)
    axes[1].set_title('Month-wise Box Plot (Seasonality)')
    axes[1].set_xlabel('Year')
    axes[1].set_ylabel('Page Views')


    # Ajustar diseÃ±o
    plt.tight_layout()

    # Guardar imagen
    fig.savefig('box_plot.png')
    return fig
    
draw_box_plot()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv('fcc-forum-pageviews.csv', parse_dates=['date'], index_col='date')

# Limpiar datos (opcional, ajustar si es necesario)
df = df[(df['value'] >= df['value'].quantile(0.025)) & (df['value'] <= df['value'].quantile(0.975))]

# Preparar datos para box plots
df_box = df.copy()
df_box.reset_index(inplace=True)
df_box['Year'] = df_box['date'].dt.year
df_box['Month'] = df_box['date'].dt.strftime('%b')

# Ordenar los meses para que inicien en Enero
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Crear figura y ejes
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# GrÃ¡fico 1: Diagrama de caja por aÃ±o
sns.boxplot(x='Year', y='value', hue='Year', data=df_box, ax=axes[0], palette='Set2', legend=False)
axs[0].set_title('Diagrama de caja anual (Tendencia)')
axs[0].set_xlabel('AÃ±o')
axs[0].set_ylabel('Visitas a la pÃ¡gina')

# GrÃ¡fico 2: Diagrama de caja por mes
sns.boxplot(x='Month', y='value', hue='Month', data=df_box, ax=axes[1], palette='Set3', order=month_order, legend=False)
axs[1].set_title('Diagrama de caja mensual (Estacionalidad)')
axs[1].set_xlabel('Mes')
axs[1].set_ylabel('Visitas a la pÃ¡gina')

# Ajustes finales
plt.tight_layout()

# Guardar la figura (como se solicita)
fig.savefig('box_plot.png')
```

```python
def draw_bar_plot():
    # Copy and modify data for monthly bar plot
    df = pd.read_csv('fcc-forum-pageviews.csv',index_col=0,parse_dates=True)

    # Clean data
    percentil_inferior = df['value'].quantile(0.025)
    percentil_superior = df['value'].quantile(0.975)
    df = df[(df['value'] >= percentil_inferior) & (df['value'] <= percentil_superior)]

    # Extraer aÃ±o y mes
    df['Year'] = df.index.year
    df['Month'] = df.index.month_name()

    # Calcular el promedio por aÃ±o y mes
    monthly_avg = df.groupby(['Year', 'Month'])['value'].mean().reset_index()

    # Pivotar para estructurar los datos para la grÃ¡fica
    data_pivot = monthly_avg.pivot(index='Year', columns='Month', values='value')

    # Ordenar los meses en el orden correcto
    month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
                   'July', 'August', 'September', 'October', 'November', 'December']
    data_pivot = data_pivot[month_order]

    # Configurar colores con Seaborn
    colors = sns.color_palette("tab10", 12)

    # Crear figura y ejes
    plt.close('all') # Cerrar figuras previas
    fig, ax = plt.subplots(figsize=(12, 8))

    # Graficar
    data_pivot.plot(kind='bar', ax=ax, color=colors)

    # Personalizar
    ax.set_title('Average Page Views per Month', fontsize=16)
    ax.set_xlabel('Years', fontsize=12)
    ax.set_ylabel('Average Page Views', fontsize=12)
    ax.legend(title='Months', fontsize=10)
    ax.grid(axis='y')

    # Guardar la imagen y retornar la figura
    fig.savefig('bar_plot.png')
    return fig

draw_bar_plot()
```
