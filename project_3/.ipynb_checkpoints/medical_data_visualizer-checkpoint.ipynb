```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Paso 1: Importar los datos
df = pd.read_csv('medical_examination.csv')

# Paso 2:  Normalizar colesterol y gl
df.loc[df['cholesterol'] == 1, 'cholesterol'] = 0
df.loc[df['cholesterol'] > 1, 'cholesterol'] = 1

# Si glucosa es 1, se convierte a 0. Si es mayor que 1, se convierte a 1.
df.loc[df['gluc'] == 1, 'gluc'] = 0
df.loc[df['gluc'] > 1, 'gluc'] = 1

# Paso 3:  Agregar columna de sobrepeso
df['height'] = (df['height'] / 100)
df['overweight'] = df['weight']/(df['height']) ** 2
df['overweight'] = np.where(df['overweight'] <= 25 ,0, 1)

# Paso 4: Crear DataFrame para el grÃ¡fico categÃ³rico usando pd.melt
df_cat  = pd.melt(df, id_vars=['cardio'], 
                value_vars=['cholesterol', 'gluc', 'alco','overweight','smoke','active'], 
                var_name='variable', value_name='value'
                )

# Paso 5: Agrupar y reformatear los datos
df_cat = df_cat.groupby(['cardio','variable','value']).size().reset_index(name='total')

# Paso 6: Renombrar columna para el grÃ¡fico
df_cat = df_cat.rename(columns={'value': 'status'})

# Paso 7: Dibujar el grÃ¡fico categÃ³rico
def draw_cat_plot():
    g = sns.catplot(
        data=df_cat,
        x='variable',
        y='total',
        hue='status',
        col='cardio',
        kind='bar',
        height=6,
        aspect=1
    )
    g.set_axis_labels('Variable', 'Total')
    plt.show()

draw_cat_plot()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Paso 1: Importar los datos
df = pd.read_csv('medical_examination.csv',index_col="id")

# Paso 2:  Normalizar colesterol y gl
df.loc[df['cholesterol'] == 1, 'cholesterol'] = 0
df.loc[df['cholesterol'] > 1, 'cholesterol'] = 1

# Si glucosa es 1, se convierte a 0. Si es mayor que 1, se convierte a 1.
df.loc[df['gluc'] == 1, 'gluc'] = 0
df.loc[df['gluc'] > 1, 'gluc'] = 1

# Paso 3:  Agregar columna de sobrepeso
df['height'] = (df['height'] / 100)
df['overweight'] = df['weight']/(df['height']) ** 2
df['overweight'] = np.where(df['overweight'] <= 25 ,0, 1)

# Paso 4: Crear DataFrame para el grÃ¡fico categÃ³rico usando pd.melt
df_cat  = pd.melt(df, id_vars=['cardio'], 
                value_vars=['cholesterol', 'gluc', 'alco','overweight','smoke','active'], 
                var_name='variable', value_name='value'
                )

# Paso 5: Agrupar y reformatear los datos
df_cat = df_cat.groupby(['cardio','variable','value']).size().reset_index(name='total')

# Paso 6: Renombrar columna para el grÃ¡fico
df_cat = df_cat.rename(columns={'value': 'status'})
    

# Paso 7: Dibujar el grÃ¡fico categÃ³rico
def draw_cat_plot():
    fig = sns.catplot(x='variable', y='total', hue='status', col='cardio', data=df_cat, kind='bar')
    
    # Ajustar el diseÃ±o
    fig.set_axis_labels('variable', 'total')
    fig.set_titles('cardio = {col_name}')
    fig.legend.set_title('value')
    
    # Guardar la figura
    fig.savefig('catplot.png')
    
    # Retornar la figura
    return fig

draw_cat_plot()

```

```python
df_heat= df.copy()
df_heat= df_heat[df_heat['ap_lo'] <= df_heat['ap_hi']]
df_heat=df_heat[df_heat['height'] >= df_heat['height'].quantile(0.025)]
df_heat=df_heat[df_heat['height'] <= df_heat['height'].quantile(0.975)]
df_heat=df_heat[df_heat['weight'] >= df_heat['weight'].quantile(0.025)]
df_heat=df_heat[df_heat['weight'] <= df_heat['weight'].quantile(0.975)]
df_heat
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la matriz de correlaciÃ³n
corr = df_heat.corr()

# Configurar la figura
plt.figure(figsize=(10, 8))
# Crear una mÃ¡scara para la parte superior
mask = np.triu(np.ones_like(corr, dtype=bool))
# Crear el mapa de calor con sns.heatmap()

sns.heatmap(corr,mask=mask,annot=True, cmap="RdBu_r", fmt='.1f', linewidths=0.5)

# Agregar tÃ­tulo
plt.title('Mapa de Calor de CorrelaciÃ³n')
```
